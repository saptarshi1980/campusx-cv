```python

sobel_vertical = np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])
sobel_vertical
```

```python
image.shape
```

```python
rows, cols = image.shape
rows, cols
```

```python
k = sobel_vertical.shape[0]
k
```

```python
output_matrix = np.zeros((rows - k + 1, cols - k + 1))
output_matrix
```

```python
for i in range(1, rows-1):
    for j in range(1, cols-1):
        region = image[i-1:i+2, j-1:j+2]
        output_matrix[i-1, j-1] = np.sum(region * sobel_vertical)
```

```python
output_matrix
```

```python
plt.figure = plt.figure(figsize = (10,5))

# original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap= 'gray')
plt.axis('off')


plt.subplot(1, 2, 2)
plt.title("Sobel Conv Edge Output")
plt.imshow(output_matrix, cmap = 'gray')
plt.axis('off')

plt.show()
```

```python
import numpy as np

def gaussian_kernel(size=3, sigma=1):
    """Creates a 2D Gaussian kernel."""
    ax = np.linspace(-(size - 1) / 2., (size - 1) / 2., size)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-0.5 * (np.square(xx) + np.square(yy)) / np.square(sigma))

    return kernel / np.sum(kernel)

# Example usage
size = 3  # Size of the kernel (3x3)
sigma = 1.0  # Standard deviation of the Gaussian

kernel = gaussian_kernel(size, sigma)
print(kernel)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image_path = '/content/dog_on_mars.jpg'

image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)




sobel_output = cv2.filter2D(image, -1, gaussian_kernel(size=3, sigma=1))
sobel_output.shape


import matplotlib.pyplot as plt

plt.figure(figsize=(11, 6))

# original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

# Sobel Conv Edge Output
plt.subplot(1, 2, 2)
plt.title("Gaussian Kernel")
plt.imshow(sobel_output, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load an image in grayscale
image_path = '/content/dog_on_mars.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Sobel edge detection (horizontal)
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

# Sobel edge detection (vertical)
sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])

# Apply the Sobel filter to the image
sobel_x_output = cv2.filter2D(image, -1, sobel_x)
sobel_y_output = cv2.filter2D(image, -1, sobel_y)

# Plotting the results
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

# Sobel X (Horizontal Edges)
plt.subplot(1, 3, 2)
plt.title("Sobel X (Horizontal Edges)")
plt.imshow(sobel_x_output, cmap='gray')
plt.axis('off')

# Sobel Y (Vertical Edges)
plt.subplot(1, 3, 3)
plt.title("Sobel Y (Vertical Edges)")
plt.imshow(sobel_y_output, cmap='gray')
plt.axis('off')

plt.show()
```
